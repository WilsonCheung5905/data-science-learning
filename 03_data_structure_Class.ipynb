## Big O andTime Complexity
https://wiki.python.org/moin/TimeComplexity

```python
# O(n)
for i in range(n):
    print(i)
    
# O(n2) square
for i in range(n):
    for j in range(n):
        print(i, j)

# O(2n)
for i in range(n):
    print(i)
for j in range(n);
    print(j)
```
